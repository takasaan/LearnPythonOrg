```python
#Classes and Objects
#Objects are an encapsulation of variables and functions into a single entity. 
#Objects get their variables and functions from classes. 
#Classes are essentially a template to create your objects.A very basic class would look something like this:

#Objectとは、変数(variables)と関数(functions)を1つにまとめた単位です。
#Objectはクラスのインスタンスであり、クラスに定義された変数(variables)と関数(functions)を継承します。
#クラスは、Objectを作成するためのテンプレートのようなものです。
#基本的なclassはこのように表記されます：

class MyClass:
    variable = "blah"#objectと呼ばれる部分
    def function(self):#objectと呼ばれる部分
        print("This is a message inside the class.")
```

```python
#We'll explain why you have to include that "self" as a parameter a little bit later. 
#First, to assign the above class(template) to an object you would do the following
#なぜ"self"をparameter(仮引数)として入れるのかは後で説明します。
#まずは1番上のclass(テンプレート)をobjectとして設定するには次のようにします。

class MyClass:
    variable = "blah"　#ここがobject

    def function(self): #ここもobject
        print("This is a message inside the class.")

myobjectx = MyClass()
#ここでMyClassの変数(variable)とfunction()が含まれた状態でインスタンス化(クラスからオブジェクトを作り出すこと)される
```

```python
#Accessing Object Variables：オブジェクトの変数にアクセスする
#To access the variable inside of the newly created object "myobjectx" you would do the following: 
#新しく作成したオブジェクト"myobjectx"内の変数(variable)にアクセスするには、次のように書きます

class MyClass:
    variable = "blah"

    def function(self):
        print("This is a message inside the class.")

myobjectx = MyClass()

myobjectx.variable
```

```python
#So for instance the below would output the string "blah":
#なので、下のinstanceでは blah が出力されます

class MyClass:
    variable = "blah"

    def function(self):
        print("This is a message inside the class.")

myobjectx = MyClass()

print(myobjectx.variable)
```

```python
#You can create multiple different objects that are of the same class(have the same variables and functions defined). 
# However, each object contains independent copies of the variables defined in the class. 
# For instance, if we were to define another object with the "MyClass" class and then change the string in the variable above:

#同じ変数(variable)と関数(function)を含むclassを使って複数の違うobjectを作ることができますが、
# それぞれのobjectが独立した変数となります。
# 例えば、別のobjectを "MyClass" class と定義した後に、変数の中身を変更しても他のobjectの変数には影響がない

class MyClass:
    variable = "blah"

    def function(self):
        print("This is a message inside the class.")

myobjectx = MyClass()
myobjecty = MyClass()

myobjecty.variable = "yackity"#ここで myobjectyの変数は"yackity"ですよと定義してあげてる

# Then print out both values
print(myobjectx.variable) #blahが出力される
print(myobjecty.variable) #yackityが出力される
```

```python
#The __init__() function, is a special function that is called when the class is being initiated. 
# It's used for assigning values in a class.
#__init__()関数はclassがインスタンス化される時、自動的に呼び出される特別なfunction(関数)です。
#classにvalues(値)をassign(代入)するときに使われます

class NumberHolder:#NumberHolderという名前のクラスを作った

   def __init__(self, number): #インスタンス化時に自動で呼び出され、numberを受け取る
       self.number = number#self.numberというインスタンス変数に、受け取ったnumberを代入

   def returnNumber(self): #returnNumberというインスタンスメソッドを作った
       return self.number#インスタンス変数self.numberを返す

var = NumberHolder(7)# NumberHolderクラスをインスタンス化し、numberに7を渡して変数varに代入
print(var.returnNumber()) #Prints '7'
```
